```python
from selenium import webdriver          #Основной элемент
from selenium.webdriver.support.ui import Select, WebDriverWait   # Select - Поле множественного выбора
import re
import requests
```

```python
link = "https://www.pik.ru/obninsk/moskovsky_kvartal/progress/3966"
driver = webdriver.Chrome()   #Движок Хрома
driver.get(link)
```

```python
def monthname_to_monthnum(monthname):
    monthnums = {
        "Январь": 1,
        "Февраль": 2,
        "Март": 3,
        "Апрель": 4,
        "Май": 5,
        "Июнь": 6,
        "Июль": 7,
        "Август": 8,
        "Сентябрь": 9,
        "Октябрь": 10,
        "Ноябрь": 11,
        "Декабрь": 12
    }
    return str(monthnums.get(monthname, monthname))
```

```python
reports_links_dict = {}
select = Select(driver.find_element_by_class_name('SelectMenu').find_element_by_tag_name("select"))
reports_months_count = len(select.options)
for i in range(reports_months_count):
    select.select_by_index(i)
    cur_report_month = select.all_selected_options[0].text
    cur_report_month = cur_report_month.split()[1] + "." + monthname_to_monthnum(cur_report_month.split()[0][:-1])
    print("Отчет за " + cur_report_month)
    reports_links_dict[cur_report_month] = {}
    for imageBlock in driver.find_elements_by_class_name("ComplexProgressPopup-imageBlock"):
        imageBlock_date = imageBlock.find_element_by_class_name("ComplexProgressPopup-imageTitle").text
        print("\tДата:", imageBlock_date)
        reports_links_dict[cur_report_month][imageBlock_date] = []
        for imageBlock_imglink in imageBlock.find_elements_by_class_name("ComplexProgressPopup-item"):
            link = "https://" + re.findall("0.news.*.jp[e]*g", imageBlock_imglink.get_attribute("style"))[0]
            reports_links_dict[cur_report_month][imageBlock_date].append(link)
            print("\t\t", link)
```

```python
for yearmonth in reports_links_dict:
    os.makedirs(os.path.join(outputdir, yearmonth), exist_ok=True)
    for dateinmonth in reports_links_dict[yearmonth]:
        tmp_outputdir = os.path.join(outputdir, yearmonth, dateinmonth)
        os.makedirs(tmp_outputdir, exist_ok=True)
        for link in reports_links_dict[yearmonth][dateinmonth]:
            tmp_outputfile = os.path.join(tmp_outputdir, os.path.basename(link))
            if not os.path.isfile(tmp_outputfile):
                img_data = requests.get(link).content
                with open(tmp_outputfile, 'wb') as handler:
                    handler.write(img_data)
```
